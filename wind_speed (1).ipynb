```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv("wind_dataset.csv", encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#check the null values
df.isnull().sum()
```

```python
#remove null values
df.dropna(inplace=True)
```

```python
#check if the null values are removed
df.isnull().sum()
```

```python
sns.histplot(df["Average wind speed [knots]"])
plt.plot()
```

```python
#check all unique values of Precipitation Amount (mm)
df["Precipitation Amount (mm)"].unique()
```

```python
#check all punique values of DATE
df["DATE"].unique()
```

```python
#check all unique values of Maximum Temperature (°C)
df["Maximum Temperature (°C)"].unique()
```

```python
#import LableEncoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['DATE'] = label_encoder.fit_transform(df['DATE'])
```

```python
df.head()
```

```python
#separate x and y
x = df.drop("Average wind speed [knots]",axis=1)
y = df["Average wind speed [knots]"]
```

```python
#split the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#build the LinearRegression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#Make predictions
ypred = model.predict(x_test)
```

```python
#Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, ypred)   #error
mse = mean_squared_error(y_test, ypred)    #error
r2 = r2_score(y_test, ypred)   #accuracy
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```
